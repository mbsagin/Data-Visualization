```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,3,9]
z = [5,2,7]

plt.plot(x,y)
plt.plot(x,z)
plt.title("TITLE FOR MY PLOT")
plt.xlabel("X LABEL")
plt.ylabel("Y LABEL")
plt.legend(["THIS IS Y", "THIS IS Z"])
plt.show()
```

```python
sample_data = pd.read_csv('sample_data.csv')
plt.plot(sample_data.column_a, sample_data.column_b)
plt.show()
```

```python
data = pd.read_csv('countries.csv')
uk = data[data.country == "United Kingdom"]
ger = data[data.country == "Germany"]
tr = data[data.country == "Turkey"]
plt.plot(uk.year, uk.population / 10**6)
plt.plot(ger.year, ger.population / 10**6)
plt.plot(tr.year, tr.population / 10**6)
plt.title("United Kingdom, Germany and Turkey Pop. Graph")
plt.xlabel("YEAR")
plt.ylabel("MILLION")
plt.legend(["UK", "GER", "TUR"])
plt.show()
```
